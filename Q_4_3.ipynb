### Inverse of the matrix

$$ A =
\begin{bmatrix}
2 & 1 & 3 \\
1 & 0 & 2 \\
3 & 4 & 1
\end{bmatrix}
$$

```python
import numpy as np
```

```python
A = np.array([[2,1,0,3,4],[1,2,1,0,2],[0,1,3,2,1],[4,0,2,0,3],[1,3,2,0,1]])
```

```python
row_3 = A[3].copy()
row_4 = A[4].copy()
A[4] = row_3; A[3] = row_4
```

```python
# A
```

```python
n = len(A)
aux = np.hstack((A, np.eye(n)))
```

```python
for i in range(n):
#     making diagonal elements 1
    pivot = aux[i][i]
    aux[i] /= pivot
    
#     making other elements zero
    for j in range(n):
        if j!=i:
            factor = aux[j, i]
            aux[j] -= factor * aux[i]

inv_A = aux[:, n:]
```

```python
inv_A
```
