### Solve it (Newton Raphson methode)
$$x^2 - 48 = 0$$

```python
import numpy as np
```

```python
def f(x):
    return x**2 - 48

def df(x):
    return 2*x
```

```python
def solution(n=1000):
    x0 = 1
    for i in range(n):
        x = x0 - f(x0)/df(x0)
        if abs(x0-x)<0.0001:
            break
        else:
            x0 = x
    print(f"root: {x} on iteration {i}")
```

```python
solution()
```
