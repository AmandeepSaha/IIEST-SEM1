## Solve this (using rk4 method)

### $$\frac{dx}{dt} = 5x - 3y \\ \frac{dy}{dt} =  âˆ’6x + 2y$$


### Initial Condition $x(0) = 15 & y(0) = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def S(t, x, y):
    return [
        5*x - 3*y,
        -6*x + 2*y
    ]
```

```python
def Solution(n=1000):
#     for saving datapoints
    configuration_space = {}
    time = []; X = []; Y = []
#--------------------------------------------------------
#     initialization
    t = 0
    x_0 = 15
    y_0 = 0
#--------------------------------------------------------
#     boundaries
    a = 0; b = 30
#--------------------------------------------------------
#     steps
    h = np.abs(b-a)/n
#--------------------------------------------------------
#     rk4 
    for i in range(n):
        k1x, k1y = S(t, x_0, y_0)
        
        k2x, k2y = S(t+h/2, x_0+(k1x*h/2), y_0+(k1y*h/2))
        
        k3x, k3y = S(t+h/2, x_0+(k2x*h/2), y_0+(k2y*h/2))
        
        k4x, k4y = S(t+h, x_0+(k3x*h), y_0+(k3y*h))
        
        kx, ky = (k1x + 2*k2x + 2*k3x + k4x)*(h/6), (k1y + 2*k2y + 2*k3y + k4y)*(h/6)
        
        t += h; x_0 += kx; y_0 += ky
        
        time.append(t)
        X.append(x_0)
        Y.append(y_0)
    
    configuration_space["time"] = time
    configuration_space["x"] = X
    configuration_space["y"] = Y
# -------------------------------------------------------
#     result
    return configuration_space
```

```python
# instance
points = Solution()
t, x, y = points["time"], points["x"], points["y"]
```

```python
# plots
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t, x, y)
ax.set_xlabel("time axis")
ax.set_ylabel("x")
ax.set_zlabel("y")
```
